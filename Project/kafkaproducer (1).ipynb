```python
import json
from kafka import KafkaProducer
import pandas as pd
import time

# Set your Kafka broker connection string
bootstrap_servers = "localhost:9092"

# Set the Kafka topic to produce to
topic = "chatgpt"

# Create a Kafka producer
producer = KafkaProducer(bootstrap_servers=bootstrap_servers)

# Read the dataset
dataset = []

df= pd.read_csv("/Users/sahanaraj/Downloads/chatgpt1.csv")
dataset = df.values.tolist()

# Produce the data to the Kafka topic

for record in dataset:
    print(record[2])
    producer.send(topic, json.dumps(record[2]).encode("utf-8"))
    producer.flush()
    time.sleep(3)
```
